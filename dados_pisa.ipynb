## Autonomia Escolas

```python
# vars das perguntas de autonomia 
nomes_vars = df_escolas.filter(regex='SC202Q*', axis=1).columns.tolist()
nomes_vars
```

```python
# criando um df c apenas as informa√ß√µes de autonomia
lista_vars = ['CNTSCHID'] + nomes_vars + ['SCHAUTO', 'SRESPCUR', 'SRESPRES', 'STRATUM']
df_SCHAUTO = df_escolas[lista_vars]
df_SCHAUTO
```

```python
dict_col = {'BRA01': 'BRA - stratum 01: Norte / Public Federal ',
'BRA02': 'BRA - stratum 02: Norte / Public State ',
'BRA03': 'BRA - stratum 03: Norte / Public Municipal ',
'BRA04': 'BRA - stratum 04: Norte / Private ',
'BRA05': 'BRA - stratum 05: Nordeste / Public Federal ',
'BRA06': 'BRA - stratum 06: Nordeste / Public State ',
'BRA07': 'BRA - stratum 07: Nordeste / Public Municipal ',
'BRA08': 'BRA - stratum 08: Nordeste / Private ',
'BRA09': 'BRA - stratum 09: Sul / Public Federal ',
'BRA10': 'BRA - stratum 10: Sul / Public State ',
'BRA11': 'BRA - stratum 11: Sul / Public Municipal ',
'BRA12': 'BRA - stratum 12: Sul / Private ',
'BRA13': 'BRA - stratum 13: Sudeste / Public Federal ',
'BRA14': 'BRA - stratum 14: Sudeste / Public State ',
'BRA15': 'BRA - stratum 15: Sudeste / Public Municipal ',
'BRA16': 'BRA - stratum 16: Sudeste / Private ',
'BRA17': 'BRA - stratum 17: Centro-Oeste / Public Federal ',
'BRA18': 'BRA - stratum 18: Centro-Oeste / Public State ',
'BRA19': 'BRA - stratum 19: Centro-Oeste / Public Municipal ',
'BRA20': 'BRA - stratum 20: Centro-Oeste / Private '
}

df_SCHAUTO = df_SCHAUTO.replace(dict_col)
df_SCHAUTO['STRATUM'] = df_SCHAUTO['STRATUM'].str.split(":", n=1, expand=True)[1]
df_SCHAUTO['REGIAO'] = df_SCHAUTO['STRATUM'].str.split("/", n=1, expand=True)[0]
df_SCHAUTO['TIPO_ESCOLA'] = df_SCHAUTO['STRATUM'].str.split("/", n=1, expand=True)[1]
df_SCHAUTO
```

```python
# duplicados?
df_SCHAUTO[df_SCHAUTO['CNTSCHID'].duplicated()]
```

# Juntando dados Escola e Estudantes

```python
df_completa = pd.merge(df_pisa_br_math, df_SCHAUTO, how='left', on='CNTSCHID')
df_completa
```

```python
# m√©dia das notas por escolas
df_media_escola = df_completa.pivot_table(index='CNTSCHID',                
                            values= ['media_nota'],
                            aggfunc='mean').reset_index()
df_media_escola.rename(columns={'media_nota': 'media_por_escola'}, inplace=True)
df_media_escola
```

```python
df_completa = pd.merge(df_completa, df_media_escola, how='left', on='CNTSCHID')
df_completa.sort_values('CNTSCHID')
```

```python
# selecionando dados das escolas
col_lista = ['CNTSCHID', 'REGIAO', 'TIPO_ESCOLA', 'SCHAUTO', 'SRESPCUR', 'SRESPRES', 'media_por_escola']
df_analise_escola = df_completa[col_lista].drop_duplicates()
df_analise_escola
```

```python
# m√©tricas por Regiao e Tipo de Escola por m√©dia de nota por escola
df_pvt_escola = (df_analise_escola
                 .pivot_table(index=['REGIAO', 'TIPO_ESCOLA'],
                              values=['media_por_escola'],
                              aggfunc=['mean', 'std', 'min', 'max', 'count'])
                 .reset_index()                
)

df_pvt_escola.columns = ['REGIAO', 'TIPO_ESCOLA', 'media', 'desvio.p', 'min', 'max', 'quantidade']
df_pvt_escola.sort_values(['REGIAO', 'TIPO_ESCOLA'])
```

```python
# m√©tricas por Regiao e Tipo de Escola por m√©dia de 'SCHAUTO', 'SRESPCUR', 'SRESPRES'
df_pvt_escola_autonomia = (df_analise_escola
    .pivot_table(index=['REGIAO', 'TIPO_ESCOLA'],
                values=['SCHAUTO', 'SRESPCUR', 'SRESPRES'],
                aggfunc=['mean'])
    .reset_index()                
)
df_pvt_escola_autonomia.columns = ['REGIAO', 'TIPO_ESCOLA', 'media_auto', 'media_SRESPCUR', 'media_SRESPRES']
df_pvt_escola_autonomia
```

### Analisando os dados: Autonomia

```python
df_pvt_auto = df_pvt_escola.merge(df_pvt_escola_autonomia, how='left', on=['REGIAO', 'TIPO_ESCOLA']).sort_values(['media_auto', 'media'])
# df_pvt_auto.to_excel('dados_analise.xlsx', index=False)
df_pvt_auto
```

```python
df_pvt_tipo_escola = df_completa.pivot_table(index=['TIPO_ESCOLA'],
                        values=['media_nota'],
                        aggfunc=['mean', 'count']).reset_index()
df_pvt_tipo_escola.columns = ['TIPO_ESCOLA', 'MEDIA_NOTA', 'QTD']
df_pvt_tipo_escola.sort_values('QTD')
```

```python
df_pvt_regiao = df_completa.pivot_table(index=['REGIAO'],
                        values=['media_nota'],
                        aggfunc=['mean', 'count']).reset_index()

df_pvt_regiao.columns = ['REGIAO', 'MEDIA_NOTA', 'QTD']
df_pvt_regiao.sort_values('QTD')
```
